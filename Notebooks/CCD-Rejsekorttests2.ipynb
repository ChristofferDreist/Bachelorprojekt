```python
import pandas as pd
import re

df = path

# Split the sequences into lists
df['StopIds'] = df['StopNbSequence'].str.split(';')
df['StopNames'] = df['StopIdSequence'].str.split(';')

# Function to clean station names by removing suffixes
def clean_station_name(name):
    name = re.sub(r'\s*\(.*?\)', '', name).strip()  # Remove parentheses and their content
    name = re.sub(r'\s+St\.$', '', name)  # Remove " St." if it appears at the end
    return name

# Create a new column for cleaned station names
df['CleanStopNames'] = df['StopNames'].apply(lambda names: [clean_station_name(n) for n in names])

df.head()
```

```python
Cityringen = ["Frederiksberg Allé", "Frederiksberg", "Aksel Møllers Have", "Nuuks plads", "Nørrebros Runddel", "Nørrebro", "Skjolds Plads", "Vibenshus Runddel", "Poul Henningsens Plads", "Trianglen", "Østerport", "Marmorkirken", "Kongens Nytorv", "Gammel Strand", "Rådhuspladsen", "København H", "Enghave Plads"]



# Function to check if a station is on the Cityringen
def is_on_cityringen(stations):
    return any(name in Cityringen for name in stations)

# Create a new column to check if the station is on the Cityringen
df['IsOnCityringen'] = df['CleanStopNames'].apply(is_on_cityringen)

df.head()
```

# Now we start with subset where Cityringen is applicable

```python
# Remove the rows that are not on the Cityringen
df = df[df['IsOnCityringen']]
df.shape
df.head()
```

```python
# Function to check if a trip is relevant to Cityringen
def is_valid_cityringen_trip(stations):
    if len(stations) == 2:  # Only origin and destination, no intermediates
        return stations[0] in Cityringen or stations[-1] in Cityringen  # Either start or end must be in Cityringen
    return True  # Otherwise, valid as there are intermediate stops

# Apply the filter to mark valid trips
df['IsValidCityringenTrip'] = df['CleanStopNames'].apply(is_valid_cityringen_trip)

# Filter out the invalid trips
df = df[df['IsValidCityringenTrip']]


print(df.shape)
df.head()
```

```python
df['StartStation'] = df['CleanStopNames'].apply(lambda x: x[0])
df['EndStation'] = df['CleanStopNames'].apply(lambda x: x[-1])

# Filtrer rejser der starter og slutter udenfor Cityringen
df_outside = df[~df['StartStation'].isin(Cityringen) & ~df['EndStation'].isin(Cityringen)]

print(f"Antal rejser hvor start/slut er udenfor Cityringen: {len(df_outside)}")
df_outside.head()
```

```python
# Opdateret funktion til at finde ruter, der bruger Cityringen korrekt
def uses_cityringen_correctly(stops, mode_seq):
    # Tjek om ruten indeholder mindst én station fra Cityringen
    cityringen_used = any(s in Cityringen for s in stops)
    
    # Tjek om 'Metro' er i transportmidlerne
    metro_used = "Metro" in mode_seq

    # Kun hvis begge er sande, returner True
    return cityringen_used and metro_used

# Anvend funktionen på DataFrame
df['UsesCityringen'] = df.apply(lambda row: uses_cityringen_correctly(row['CleanStopNames'], row['ModeSeq']), axis=1)

# Print antal ruter der bruger Cityringen
print(f"Antal rejser der bruger Cityringen: {df['UsesCityringen'].sum()}")

# Vis eksempler på ruter der bruger Cityringen
df[df['UsesCityringen']].head()
```

```python
df_cityringen = df[df['UsesCityringen']].copy()

unique_routes = df_cityringen['CleanStopNames'].drop_duplicates()

for route in unique_routes:
    print(route)
```

```python
def could_have_used_metro_but_took_bus_or_train(stops, mode_seq):
    # Cityringen-stationer i ruten, ekskl. København H & Østerport
    cityringen_stations = [s for s in stops if s in Cityringen]
    
    # Cityringen var en reel mulighed, hvis mindst én anden Cityringen-station er i ruten
    cityringen_possible = len(cityringen_stations) > 0
    
    # Metro må IKKE være brugt
    metro_not_used = "Metro" not in mode_seq
    
    # Bus eller tog SKAL være brugt
    bus_or_train_used = any(m in mode_seq for m in ["Bus", "Tog"])
    
    return cityringen_possible and metro_not_used and bus_or_train_used

# Filtrer ruter i DataFrame
df_bus_tog_instead_of_metro = df[df.apply(lambda row: could_have_used_metro_but_took_bus_or_train(row['CleanStopNames'], row['ModeSeq']), axis=1)].copy()

# Print antal fundne ruter
print(f"Antal rejser med bus/tog, hvor Cityringen var en mulighed: {len(df_bus_tog_instead_of_metro)}")

# Vis eksempler
df_bus_tog_instead_of_metro.head()
```

```python
def could_have_used_metro_but_took_bus_or_train(stops, mode_seq):
    # Cityringen-stationer i ruten, ekskl. København H & Østerport
    cityringen_stations = [s for s in stops if s in Cityringen and s not in ["København H", "Østerport"]]
    
    # Cityringen var en reel mulighed, hvis mindst én anden Cityringen-station er i ruten
    cityringen_possible = len(cityringen_stations) > 0
    
    # Metro må IKKE være brugt
    metro_not_used = "Metro" not in mode_seq
    
    # Bus eller tog SKAL være brugt
    bus_or_train_used = any(m in mode_seq for m in ["Bus", "Tog"])
    
    return cityringen_possible and metro_not_used and bus_or_train_used

# Filtrer ruter i DataFrame
df_bus_tog_instead_of_metro = df[df.apply(lambda row: could_have_used_metro_but_took_bus_or_train(row['CleanStopNames'], row['ModeSeq']), axis=1)].copy()

# Print antal fundne ruter
print(f"Antal rejser med bus/tog, hvor Cityringen var en mulighed: {len(df_bus_tog_instead_of_metro)}")

# Vis eksempler
df_bus_tog_instead_of_metro.head()
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import accuracy_score, classification_report

# Brug de relevante kolonner
features = ['StartStation', 'EndStation', 'transfers', 'TT(min)', 'Start_Time']
target = 'UsesCityringen'

# One-Hot Encoding for kategoriske variabler (Start/End station, tid på dagen, ugedag)
encoder = OneHotEncoder(handle_unknown='ignore')
X_encoded = encoder.fit_transform(df[features]).toarray()

# Målvariabel
y = df[target].astype(int)

# Split i trænings- og testdata
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Initialiser og træn modellen
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# Forudsig testdata
y_pred = clf.predict(X_test)

# Evaluer modellen
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))

# Feature Importance
feature_importance = pd.DataFrame({'Feature': encoder.get_feature_names_out(features),
                                   'Importance': clf.feature_importances_}).sort_values(by='Importance', ascending=False)
print("Feature Importance:\n", feature_importance)
```
