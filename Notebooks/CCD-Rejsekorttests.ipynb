```python
import pandas as pd
import re

df = path

# Split the sequences into lists
df['StopIds'] = df['StopNbSequence'].str.split(';')
df['StopNames'] = df['StopIdSequence'].str.split(';')

# Function to clean station names by removing suffixes
def clean_station_name(name):
    name = re.sub(r'\s*\(.*?\)', '', name).strip()  # Remove parentheses and their content
    name = re.sub(r'\s+St\.$', '', name)  # Remove " St." if it appears at the end
    return name

# Create a new column for cleaned station names
df['CleanStopNames'] = df['StopNames'].apply(lambda names: [clean_station_name(n) for n in names])

df.head()
```

```python
Cityringen = ["Frederiksberg Allé", "Frederiksberg", "Aksel Møllers Have", "Nuuks plads", "Nørrebors Runddel", "Nørrebro Station", "Skjolds Plads", "Vibenhus Runddel", "Poul Henningsens Plads", "Trianglen", "Østerport", "Marmorkirken", "Kongens Nytorv", "Gammel Strand", "Rådhuspladsen", "København H", "Enghave Plads"]



# Function to check if a station is on the Cityringen
def is_on_cityringen(stations):
    return any(name in Cityringen for name in stations)

# Create a new column to check if the station is on the Cityringen
df['IsOnCityringen'] = df['CleanStopNames'].apply(is_on_cityringen)

df.head()
```

# Now we start with subset where Cityringen is applicable

```python
# Remove the rows that are not on the Cityringen
df = df[df['IsOnCityringen']]
df.shape
df.head()
```

```python
# Function to check if a trip is relevant to Cityringen
def is_valid_cityringen_trip(stations):
    if len(stations) == 2:  # Only origin and destination, no intermediates
        return stations[0] in Cityringen or stations[-1] in Cityringen  # Either start or end must be in Cityringen
    return True  # Otherwise, valid as there are intermediate stops

# Apply the filter to mark valid trips
df['IsValidCityringenTrip'] = df['CleanStopNames'].apply(is_valid_cityringen_trip)

# Filter out the invalid trips
df = df[df['IsValidCityringenTrip']]


print(df.shape)
df.head()
```
